### 3. Add move up method

```python
#Building a class with some attributes
class Rockets():

  #use special method in python to initialize the attributes from this class
  def __init__(self):
    self.x = 0
    self.y = 0 #first arg is always self. help us refer to the object we work with

  #add a method that will make your rocket move up by 1 unit whenever it is called.

  def move_up(self):
    self.y += 1 #it is the same as self.y = self.y + 1


```

```python
#Make spaceX rocket and make it move up by two units
SpaceX = Rockets()

print(f'Current Position of SpaceX {SpaceX.x , SpaceX.y}')

#move spaceX up by 1unit
SpaceX.move_up()
print(f'Position of SpaceX updated {SpaceX.x , SpaceX.y}')

#move spaceX one more unit up
SpaceX.move_up()
print(f'Position of SpaceX updated after 1 unit {SpaceX.x , SpaceX.y}')
```

### 6 Adding Distance between Rockets

```python
#Building a class with some attributes
class Rockets():

  #use special method in python to initialize the attributes from this class
  def __init__(self, x=0, y=0):
    self.x = x
    self.y = y #first arg is always self. help us refer to the object we work with

  #add a method that will make your rocket move in either up, down, left, right in any units.

  def move_rocket(self, x_increment=0, y_increment=1):
    self.x += x_increment # increase units on the x-axis
    self.y += y_increment # increase units on the y-axis

 #Calculate Euclidian distance between two rockets
  def get_distance(self , other_rocket):
    x_diff = self.x - other_rocket.x
    y_diff = self.y - other_rocket.y
    distance = (x_diff ** 2 + y_diff ** 2) ** 0.5 ## ** squares the difference in x and y and root it by 0.5.
    return round (distance, 3) #round the answer of distance to 3 decimal places
```

```python
rocket4 = Rockets(-10 , 0)
rocket5 = Rockets(0, 0)
print (f'{rocket4.x , rocket4.y}')
print (f'{rocket5.x , rocket5.y}')
```

```python
#Distance of rocket4  from rocket5
distance_for_4_5 = rocket5.get_distance(rocket4) #rocket5.get_distance is the self and the rocket4 is the other_rocket.
print (f'The distance between rocket4 from rocket5 is:- {distance_for_4_5} (3dp)')
```

```python
# Move rocket4 to (-15, 10)
rocket4.move_rocket(-5, 10)

#Move rocket5 to (0 , 15)
rocket5.move_rocket(0 , 15)

#Get the new locations of rocket4 and rocket5
print (f'{rocket4.x , rocket4.y}')
print (f'{rocket5.x , rocket5.y}')
```
