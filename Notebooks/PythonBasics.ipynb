```python
#create a function that will either add, subtract, multiply or divide two input numbers
def Calcus(n1,n2,operation):
  if operation.lower() == 'add':
    return n1 + n2
  elif operation.lower() == 'subtract':
    return n1 - n2
  elif operation.lower() == 'multiply':
    return n1 * n2
  elif operation.lower() == 'divide':
    return n1/n2
  else:
    print(f'operation input can either take:- add, multiply, subtract or divide. you gave :- {operation}. try again')
```

```python
addition = Calcus(2,3,'add')
print (addition)
```

```python
subtraction = Calcus (2,3, 'subtract')
print (subtraction)
```

```python
Multiplication = Calcus (2,3, 'multiply')
print (Multiplication)
```

```python
division = Calcus (2,3, 'divide')
print (division)
```

```python
result = Calcus (2,3 , 'tim')
print (result)
```
